```python
import os, sqlite3, xlrd, time, datetime
import pandas as pd
from utils_query import *
```

```python
xls_file = pd.ExcelFile('query2.xlsx')
table = xls_file.parse('Sheet1')
chrom = [x for x in table['Chr'].unique()[:-1]]
conn = sqlite3.connect('dbase_Sqlite')
# cleardb('dbase_Sqlite', 'dataFrame2')
table.to_sql('dataFrame2', conn)
```

```python
startTime = time.time()
query = """
        SELECT DISTINCT
            dataFrame2.*, Indian_ALLELE_FREQ_1, Indian_ALLELE_FREQ_2, Indian.ID
        FROM
            dataFrame2 LEFT OUTER JOIN 
            Indian ON dataFrame2.Coordinate=Indian.POS 
            AND dataFrame2.Chr=Indian.CHROM
        """
df_I = pd.read_sql_query(query , conn)

query = """
        SELECT DISTINCT
            dataFrame2.Coordinate, Malay_ALLELE_FREQ_1, Malay_ALLELE_FREQ_2, Malay.ID
        FROM
            dataFrame2 LEFT OUTER JOIN
            Malay ON dataFrame2.Coordinate=Malay.POS
            AND dataFrame2.Chr=Malay.CHROM
        """
df_M = pd.read_sql_query(query , conn)

query = """
        SELECT DISTINCT
            dataFrame2.Coordinate, Chinese_ALLELE_FREQ_1, Chinese_ALLELE_FREQ_2, Chinese_ALLELE_FREQ_3, Chinese_ALLELE_FREQ_4, Chinese.ID
        FROM
            dataFrame2 LEFT OUTER JOIN
            Chinese ON dataFrame2.Coordinate=Chinese.POS
            AND dataFrame2.Chr=Chinese.CHROM
        """

df_C = pd.read_sql_query(query , conn)
        
timeTaken = time.time()-startTime 
print(datetime.datetime.now().strftime('%Y/%m/%d %H:%M:%S'), 
      '- Data Retrieval for all races completed: Took {} seconds to complete.'.format(timeTaken))
```

```python
df = df_I.merge(df_C, on=['Coordinate'], how='outer')
df
```

```python
df_I.to_excel("output_I.xls")
df_M.to_excel("output_M.xls")
df_C.to_excel("output_C.xls")
# df.to_excel("output.xls")
```
